## 1ï¸âƒ£ Cast a Spell (Pure String â†’ JIT Kernel)

```python
spell = "(ğ•‚ğŸ›ğ”» âŠ— â„¤ğ•Šğ•â„ âŠ— ğ”¾â„ğ”½)(beta=0.8, sigma=1.2, rank=64)"

cell, state = run_spell_cell(
    spell,
    steps=1,
    seed=42,
    viz=True,
    fmt="svg"
)

print("âœ… Spell cast â€“ final free-energy:", float(state.free_energy))
```

## 2ï¸âƒ£ Live Telemetry Dashboard (From Cell History)

```python
# Run more steps with live update
cell2, state2 = run_spell_cell(spell, steps=0, seed=43)  # init only

key = jax.random.PRNGKey(0)
for i in range(100):
    key, subkey = jax.random.split(key)
    cell2.step(subkey)
    
    # Simple live telemetry
    clear_output(wait=True)
    fe = float(cell2.state.free_energy) if hasattr(cell2.state, 'free_energy') else 'N/A'
    x_mean = cell2.history[-1].x_mean if cell2.history else 0.0
    print(f"Step {i+1:3d} | Î¦={x_mean:.4f} | â„±={fe}")
```
